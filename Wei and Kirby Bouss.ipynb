```python
from sympy import *
```

```python
u,A,B,x,t,C,h,A1,A2,z,k,h,H,mu,delta,alpha,z_alpha,C2 = symbols('u,A,B,x,t,C,h,A1,z,A2,k,h,H,mu,delta,alpha,z_alpha,C2')
k = 2.0*pi/1.0
h = 0.5
z_alpha = 0.531*h
H =0.1*h
alpha = (z_alpha**2)/2.0 + z_alpha
delta = H/h
mu = k*h
f = 2.0*alpha*C2**3 - (3.0*alpha + 1.0/3.0 + 2.0*alpha*delta)*C2**2 + 2.0*delta*(alpha+1.0/3.0)*C2+alpha + 1.0/3.0
sol = solve(f,C2)
for c in sol:
    print c.n()
    print re(c.n())
C = sqrt(re(sol[1].n()))
A = (C**2 - 1) / (delta * C)
B = sqrt( (C**2 - 1.0) / (4.0*mu**2*((alpha+1.0/3.0) - alpha*C**2)) )
A1 =    (C**2 - 1.0) / (3.0*delta*((alpha+1.0/3.0) - alpha*C**2))
A2 =  - ((C**2 - 1)**2 * ((alpha+1.0/3.0) + 2.0*alpha*C**2)) / (2.0*delta*C**2 * ((alpha+1.0/3.0) - alpha*C**2))
u = A/(cosh(B*(x - C*t))**2)
eta = A1/cosh(B*(x - C*t))**2 + A2/cosh(B*(x - C*t))**4
print A,B,A1,A2
```

```python
dX = 25./100.
X = [float(i*dX) for i in range(101)]
UL = [u.evalf(subs={t:2.0,x:xi}) for xi in X]
HL = [h+H*eta.evalf(subs={t:2.0,x:xi}) for xi in X]
pyplot.plot(X,HL)
```

```python
HL = [h+eta.evalf(subs={t:5.0,x:xi}) for xi in X]
UL = [u.evalf(subs={t:5.0,x:xi}) for xi in X]
pyplot.plot(X,HL)
```
