```python
%matplotlib inline
from matplotlib import pyplot as plt
from proteus.WaveTools import RandomWaves, MultiSpectraRandomWaves
import numpy as np

n = 300

eta_t = np.zeros(n)

Nspectra = 2
Tp = [2., 2.]
Hs = [0.15, 0.3]
depth = 1.
mwl = [0., 0.]
g = np.array([0., 0., -9.81])
N = [101, 101]
bandFactor = [2., 2.]
spectName = ["JONSWAP", "JONSWAP"]
spectral_params = [None, None]
phi = [None, None]
waveDir = [ [1., 2., 0.], [1., 1., 0.] ]



MS = MultiSpectraRandomWaves(Nspectra,
                             Tp,
                             Hs,
                             mwl,
                             depth,
                             waveDir,
                             g,
                             N,
                             bandFactor,
                             spectName,
                             spectral_params,
                             phi
                             )

tnlist=np.linspace(0, 40, n)

for i in range(len(tnlist)):
    x = [2., 0.5, 0.]
    t = tnlist[i]
    eta_t[i] = MS.eta(x, t)
    

plt.figure(figsize=(10,6), dpi=100)

plt.xlim(0., 40.)
plt.plot(tnlist, eta_t)

plt.xlabel("Time (s)", size=20)
plt.ylabel("$\eta$ (m)", size=20)
plt.grid()
plt.show()
```
