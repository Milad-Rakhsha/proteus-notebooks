```python
%pylab inline
h = 0.5*0.61
H = 0.3*h
g = 9.8
c = sqrt(g*(h+H))
def eta(x,t):
    return H/cosh(sqrt(3.0*H/(4.0*h**3)) * (x - c*t))**2
def u(x,t):
    return c*eta(x,t)/(eta(x,t)+h)
L = 5.0
nx=101
dx = L/float(nx-1)
xl = [i*dx for i in range(nx)]
hl = [h+eta(x,1.0) for x in xl]
ul = [u(x,1.0) for x in xl]
print max(hl)-h
#pyplot.axis('equal')
pyplot.plot(xl,hl)
```

```python
pyplot.plot(xl,ul)
```

```python
def waveHeight(x,t):
    return h + eta(x,t)
def wave_u(x,t):
    return c*(waveHeight(x,t)-h)/waveHeight(x,t)
xl = [i*dx for i in range(nx)]
hl = [waveHeight(x,1.0) for x in xl]
ul = [wave_u(x,1.0) for x in xl]
pyplot.plot(xl,hl)
```
